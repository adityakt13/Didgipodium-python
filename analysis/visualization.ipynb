```python
tips = px.data.tips()
tips
```

```python
tips.sex.value_counts()
```

```python
tips.sex.value_counts().plot(kind='pie',
                            autopct='%.1f%%', 
                            colors=['black','red'],
                            textprops={'color':'w', 
                                       'fontsize':20},
                            wedgeprops={'width':0.5},
                            pctdistance=0.7,
                            )
```

```python
tips.day.value_counts().plot(kind='bar',figsize = (6,4), title = 'Day wise comparison')
plt.savefig('daywise.png', dpi = 300 , bbox_inches = 'tight', pad_inches = 2)
```

```python
tips.day.value_counts().plot(
    kind = 'bar',
    figsize = (6,4),
    hatch = '//',
    edgecolor = 'white',
    title = 'Day wise comparison'
)
```

```python
ax = tips.total_bill.plot(kind = 'hist',
                    figsize = (6,4),
                    title = 'Total Bill Distribution',
                    bins = 20,
                    color = 'red',
                    edgecolor = 'black',
                    alpha = .5)
```

```python
tips.plot(kind= 'box', x = 'day',
          y = 'total_bill', figsize = (6,4) ,
          notch = True )
```

```python
tips.total_bill.plot.box(
    figsize = (6,4),
    showmeans = True,
    meanline = True,
    meanprops = {'color':'green', 'linewidth': .5, 'linestyle': '--'},
    notch = True,
)
plt.show()
```

```python
tips.boxplot(column = 'total_bill', by = 'day', figsize =(6,4))
```

```python
px.box(tips,x = 'day',y = 'total_bill')
```

```python
px.box(tips, x='sex', y='tip')
```
